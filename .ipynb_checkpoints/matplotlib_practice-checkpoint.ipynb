```python
# Add dependencies
import pandas as pd
import os 
import numpy as np 
import matplotlib.pyplot as plt

%matplotlib inline

#MATLOAB approach: anything that starts with "plt.""
#OOM approach: starts with "ax.""
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
#MATLAB method
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
#Object Oriented method
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
#creats an empty figure
fig = plt.figure()
ax = fig.add_subplot()
```

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the plot.
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
```

```python
# Create the plot with ax.plt()
#Object Oriented method
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis, color="green", marker="^", label='Boston')
ax.set_xlabel('Date')
ax.set_ylabel('Fares($)')
ax.set_ylim(0, 45)
ax.set_title("PyBer Fare by Month")
ax.legend()
ax.grid()
```

```python
# Create the plot for a bar chart (MATLAB)
plt.bar(x_axis, y_axis)
```

```python
#Annotated bar chart (MATLAB)
# Create the plot.
plt.bar(x_axis, y_axis, color="green", label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the plot (MATLAB)
plt.barh(x_axis, y_axis)
```

```python
#Annotated bar chart (MATLAB)

plt.barh(x_axis, y_axis,color="magenta", label='Boston')
plt.gca().invert_yaxis()
plt.xlabel("Date")
plt.ylabel("Fare($)")
plt.title("PyBer Fare by Month")
plt.legend()
```

```python
plt.scatter(x_axis, y_axis)
```

```python
y_axis_larger = []
for data in y_axis:
  y_axis_larger.append(data*3)

plt.scatter(x_axis, y_axis, s=  y_axis_larger)
```

```python
#using list comprehension
#List comprehension is faster because it is optimized for Python to spot a predictable pattern during looping.

plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

```python
#OOM
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, label= "Boston",s = [i * 5 for i in y_axis], c="skyblue", alpha=.8, edgecolors="black", linewidths=2)
ax.invert_yaxis()
ax.set_xlabel("Date")
ax.set_ylabel("Fare($)")
ax.set_title("PyBer Fare by Month")
ax.set_xlim(0,50)

ax.legend()
```

```python
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
plt.subplots(figsize=(8, 8))
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

#plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8, 8))

colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)

ax.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    shadow=True,
    counterclock=False,
    startangle=95,
    autopct='%.1f%%')

plt.show()
```
